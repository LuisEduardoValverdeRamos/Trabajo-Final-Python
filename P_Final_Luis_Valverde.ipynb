```python
import pandas as pd
import numpy as np
import altair as alt


dt = "https://github.com/LuisEduardoValverdeRamos/Trabajo-Final-Python/raw/main/IDH_PNUD_2017.xlsx"
data_cruda= pd.ExcelFile(dt)
data_cruda2 = data_cruda.sheet_names
```

```python
data_cruda3 = pd.read_excel(data_cruda, sheet_name='Provincial')
data_cruda3
```

```python
variables = ["Índice de Desarrollo Humano provincial, 2017", "Unnamed: 2", "Unnamed: 5", "Unnamed: 7", "Unnamed: 9", "Unnamed: 11", "Unnamed: 13", "Unnamed: 15"]
data_cruda4 = data_cruda3[variables]
data_cruda4
```

```python
data_cruda5 = data_cruda4.dropna()
data_cruda5
```

```python
data_cruda6 = data_cruda5.reset_index(drop=True)
data_cruda7 = data_cruda6.drop(0)
```

```python
data = data_cruda7.reset_index(drop=True)
VariablesOF = ["ubigeo", "provincia", "población", "idh","e_vida", "e_com", "años", "ing"]
data.columns = VariablesOF
data
```

```python
#Gráfico Nº1: Gráfico de barras IDH según provincia
g1= alt.Chart(data).mark_bar().encode(
  y = alt.Y("idh"),
  x = alt.X("provincia")
  )

g1
```

```python
#Gráfico Nº2: Gráfico de dispersion entre nivel de ingresos y la esperanza de vida
g2 = alt.Chart(data).mark_point().encode(
    x = alt.X("ing"),
        y = alt.Y("e_vida"),
    )

g2
```

```python
#Gráfico Nº3: Gráfico de dispersion entre nivel de ingresos y la esperanza de vida
g3 = alt.Chart(data).mark_point().encode(
    x = alt.X("e_com"),
        y = alt.Y("e_vida"),
    )

g3
```

```python
#Gráfico Nº4: Gráfico de barras de ingresos según provincia

g4= alt.Chart(data).mark_bar().encode(
  y = alt.Y("ing"),
  x = alt.X("provincia")
  )

g4
```
